```python
def sum_lst(lst):
    # Выводим текущее значение lst
    print(lst)
    # Задаём условие выхода из рекурсии
    if len(lst) == 0:return 0
    # Во всех других случаях возвращаем
    # сумму первого элемента списка 
    # и результат суммирования оставшихся
    return lst[0] + sum_lst(lst[1:]) #происходит повторный вызов sum_lst (будто вложенный) с измененным кол-вом элементов и тд
 
my_lst = [12, 21, 24, 12]
print(sum_lst(my_lst))
print(sum(my_lst))
#Список my_lst() поступает как входные данные в функцию sum_lst.
```

```python
def fib(n):
    if n == 0: return 0
    if n == 1: return 1
    return fib(n-1)+fib(n-2)

#fib(0) = 0
#fib(1) = 1
print(fib(0))
print(fib(2))
print(fib(12))
```

```python
def power(val, n):
   if n == 0: return 1
   if n == 1: return val
   return power(val, n - 1) * power(val, 1)
print(power(25,0))
print(power(-5,4))
# 1
# 625
```
